```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Generate synthetic data
np.random.seed(42)
X = 2 * np.random.rand(100, 1)  # Predictor variable
y = 4 + 3 * X + np.random.randn(100, 1)  # Target variable with some noise

# Train a simple linear regression model
model = LinearRegression()
model.fit(X, y)

# Predictions
y_pred = model.predict(X)

# R-squared score
r2 = r2_score(y, y_pred)

# Print model parameters
print(f"Intercept: {model.intercept_[0]:.2f}")
print(f"Coefficient: {model.coef_[0][0]:.2f}")
print(f"R-squared: {r2:.2f}")

# Plot data and regression line
plt.scatter(X, y, color="blue", label="Actual Data")
plt.plot(X, y_pred, color="red", linewidth=2, label="Regression Line")
plt.xlabel("Predictor (X)")
plt.ylabel("Target (y)")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Step 1: Create Data (Grades & Pocket Money)
X = np.array([50, 60, 70, 80, 90]).reshape(-1, 1)  # Grades (Independent Variable)
y = np.array([100, 150, 200, 250, 300])  # Pocket Money (Dependent Variable)

# Step 2: Train the Linear Regression Model
model = LinearRegression()
model.fit(X, y)

# Step 3: Make Predictions
X_line = np.linspace(40, 100, 100).reshape(-1, 1)  # Generate X values for the line
y_line = model.predict(X_line)  # Predict corresponding y values

# Step 4: Plot the Data Points
plt.scatter(X, y, color="blue", label="Actual Data", marker='o', s=100)  # Data points

# Step 5: Plot the Regression Line
plt.plot(X_line, y_line, color="red", label="Regression Line", linewidth=2)

# Step 6: Labels and Title
plt.xlabel("Grade (%)")
plt.ylabel("Pocket Money (₹)")
plt.title("Linear Regression: Pocket Money vs Grades")
plt.legend()
plt.grid(True)

# Show the Plot
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Step 1: Create a Dataset (Size, Bedrooms, Age, and House Price)
data = {
    "Size (sq ft)": [1400, 1600, 1700, 1875, 1100],  # Feature 1
    "Bedrooms": [3, 4, 3, 5, 2],  # Feature 2
    "Age (years)": [10, 15, 5, 20, 25],  # Feature 3
    "Price (₹ Lakhs)": [150, 180, 175, 210, 130]  # Target variable
}

# Convert dictionary to DataFrame
df = pd.DataFrame(data)

# Step 2: Define Features (X) and Target Variable (y)
X = df[["Size (sq ft)", "Bedrooms", "Age (years)"]]  # Multiple features
y = df["Price (₹ Lakhs)"]  # Target variable

# Step 3: Train the Multiple Linear Regression Model
model = LinearRegression()
model.fit(X, y)

# Step 4: Make Predictions
y_pred = model.predict(X)

# Step 5: Model Performance (R-squared)
r2 = r2_score(y, y_pred)

# Print results
print("Predicted House Prices:", y_pred)
print("R-squared Score:", r2)

# Step 6: Visualizing Actual vs Predicted Prices
plt.scatter(y, y_pred, color="blue", label="Predicted vs Actual")
plt.plot(y, y, color="red", linestyle="dashed", label="Perfect Fit Line")
plt.xlabel("Actual House Price (₹ Lakhs)")
plt.ylabel("Predicted House Price (₹ Lakhs)")
plt.title("Actual vs Predicted House Prices")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error

# Step 1: Create Data (Mileage & Car Price)
X = np.array([10000, 25000, 40000, 55000, 70000, 85000, 100000]).reshape(-1, 1) 
y = np.array([500000, 420000, 350000, 290000, 240000, 190000, 150000])  


model = LinearRegression()
model.fit(X, y)


y_pred = model.predict(X)


r2 = r2_score(y, y_pred)  
rmse = np.sqrt(mean_squared_error(y, y_pred))  


print(f"R-squared (R²): {r2:.4f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")


plt.scatter(X, y, color="blue", label="Actual Data", s=100)
plt.plot(X, y_pred, color="red", label="Regression Line", linewidth=2)
plt.xlabel("Mileage (km)")
plt.ylabel("Car Price (₹)")
plt.title("Car Price Prediction: Actual vs Predicted")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression


data = {
    "Size_sqft": [1000, 1500, 2000, 2500, 3000],
    "Bedrooms": [2, 3, 3, 4, 5],
    "Age_years": [10, 5, 2, 8, 1],
    "Price": [300000, 400000, 500000, 600000, 700000]
}

df = pd.DataFrame(data)


X = df[["Size_sqft", "Bedrooms", "Age_years"]]
y = df["Price"]


model = LinearRegression()
model.fit(X, y)


coefficients = pd.DataFrame(model.coef_, X.columns, columns=["Impact"])


print("Feature Impact on House Price Prediction:")
print(coefficients)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures


np.random.seed(0)
X = np.linspace(1, 10, 10).reshape(-1, 1) 
y = X**2 + np.random.normal(0, 5, size=X.shape)  


linear_model = LinearRegression()
linear_model.fit(X, y)
y_linear_pred = linear_model.predict(X)


poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)
poly_model = LinearRegression()
poly_model.fit(X_poly, y)
y_poly_pred = poly_model.predict(X_poly)


plt.scatter(X, y, color="blue", label="Actual Data")
plt.plot(X, y_linear_pred, color="red", label="Linear Regression (Straight Line)")
plt.plot(X, y_poly_pred, color="green", label="Polynomial Regression (Curved Line)")
plt.legend()
plt.xlabel("Speed (km/h)")
plt.ylabel("Fuel Efficiency (km/l)")
plt.title("Linear vs. Polynomial Regression")
plt.show()
```
